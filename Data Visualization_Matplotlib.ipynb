```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv(r"C:\Users\rumjhum\Desktop\Data Visualization through Matplotlib\Mattplotlib Test Data.csv")
df.dropna(axis = 1,inplace = True)
df.head()
```

```python
plt.plot([1,3,2,4,1])
plt.show()
```

```python
plt.scatter(x = df['math score'], y = df['writing score'])
plt.show()
```

```python
plt.hist(df['reading score'])
```

```python
plt.hist(df['reading score'], bins = 5)
plt.show()
```

```python
plt.boxplot(df['writing score'])
plt.show()
```

```python
plt.scatter(x = df['math score'], y = df['writing score'], color = 'yellow')
plt.xlabel('Score in Math')
plt.ylabel('Score in Writing')
plt.title('Scatter plot between Two scores')
plt.show()
```

```python
plt.figure(figsize=(16,7))
plt.scatter(x = df['math score'], y = df['writing score'], color = 'red')
plt.xlabel('Score in Math')
plt.ylabel('Score in Writing')
plt.title('Scatter plot between Two scores')
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.scatter(x = df['math score'], y = df['writing score'])
plt.scatter(x = df['math score'], y = df['reading score'])
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.scatter(x = df['math score'], y = df['writing score'], alpha = 0.5)
plt.scatter(x = df['math score'], y = df['reading score'], alpha = 0.5)
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.scatter(x = df['math score'], y = df['writing score'], alpha = 0.5, label = 'Comparison of Math and Writing Score')
plt.scatter(x = df['math score'], y = df['reading score'], alpha = 0.5, label = 'Comparison of Math and Reading Score')
plt.legend()
plt.show()
```
