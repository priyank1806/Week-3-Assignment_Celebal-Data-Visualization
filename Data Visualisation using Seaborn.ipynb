```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set_style('darkgrid')
df = pd.read_csv(r"C:\Users\rumjhum\Desktop\Data Visualization through Seaborn\Data Visualization through Seaborn.csv")
```

```python
df.head()
```

```python
plt.figure(figsize = (10,6))  # for figure size, Hue for differencing between variables
sns.scatterplot(x = df['Primary_Only'], y = df['HrSec_Only'], hue = df["year"], alpha = 0.8) # alpha for visibility of plots 
plt.show()
```

```python
plt.figure(figsize = (9,6))
sns.countplot(x = "year", data = df)
plt.show()
```

```python
plt.figure(figsize = (9,6))
sns.distplot(df['Primary_Only'], kde = False)
plt.show()
```

```python
plt.figure(figsize = (9,6))
sns.kdeplot(df['Primary_Only'])
plt.show()
```

```python
plt.figure(figsize = (9,6))
sns.regplot(x = df['Primary_Only'], 
            y = df['Sec_with_HrSec.'],
            scatter_kws = {'color':'green'},
            line_kws = {'color':'black'}
)
```

```python
plt.figure(figsize = (9,6))
sns.lmplot(x = 'Primary_Only', 
           y = 'Sec_with_HrSec.',
           hue = 'year',
           data = df
)
```

```python
sns.pairplot(df[['Primary_Only', 'Sec_with_HrSec.', 'U_Primary_With_Sec']])
```

```python
sns.jointplot(x = 'Primary_Only', 
              y = 'Sec_with_HrSec.',
              data = df
             
)
plt.show()
```

```python
sns.boxplot(x = 'Primary_Only', 
            data = df
)
plt.show()
```

```python
plt.figure(figsize = (25,6))
sns.boxplot(data = df)
plt.show()
```

```python
plt.figure(figsize = (20,8))
sns.swarmplot(x = 'Primary_Only',
              y = 'U_Primary_Only', 
              data = df,
              alpha = 0.8
)
plt.show()
```

```python
numeric_df = df.select_dtypes(include=['number'])

# Now plot the heatmap
sns.heatmap(numeric_df.corr(), annot=True, cmap='inferno')
plt.show()
```

```python
plt.figure(figsize = (20,8))
colors = [
           '#F8D030',
           '#E0C068',
           '#EE99AC',
           '#C03028',
           '#F85888'
]
sns.stripplot(x = 'State_UT',
              data = df,
              palette = colors
)
plt.show()
```
